#Most Productive teams and his Key Player in 2020 in top 5 League Europe

The 2020-2021 season has ended in each league. the season is running competitively. A lot of great and interesting things happened in that season. But I'm curious, can we see who is the most productive team at the end of the season by looking at their performances at the beginning of the season? this time i will try to see this point of view with the dataset i found from kaggle. This dataset contains teams from the top leagues in the world. I will see it in the top 5 European leagues which we know are the best leagues in the world. This dataset in 2020 was created when the leagues were still running around 9-11 gameweek

#Import Packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#Loading Data

```python
df = pd.read_csv('/content/Dataa.csv')
df
```

```python
df.info
```

```python
df.describe
```

```python
df.columns
```

```python
df['Goals_Conversion']=df['Goals']/df['OnTarget']
df['Goals_per_Match_Ratio']=df['Goals']/df['Matches_Played']
df['Minutes_per_Goals']=df['Mins']/df['Goals']
df['Diff_GoalsXG']=df['Goals']-df['xG']
df
```

```python
df2020=df[df['Year']==2020]
df2020
```

```python
df2020.columns
```

```python
league2020=df2020.groupby(by=["League"])['Matches_Played',
       'Mins', 'Goals', 'xG', 'OnTarget',].sum().reset_index()
league2020
```

```python
league2020['Goals_Conversion']=league2020['Goals']/league2020['OnTarget']
league2020['Goals_per_Match_Ratio']=league2020['Goals']/league2020['Matches_Played']
league2020['Minutes_per_Goals']=league2020['Mins']/league2020['Goals']
league2020['Diff_GoalsXG']=league2020['Goals']-league2020['xG']
league2020
```

```python
league2020
```

```python
league2020.drop([1,2,5], inplace = True)
league2020
```

```python
league2020
```

#Top 5 Europe League Rank by Goals Conversion, Goals per Match, Minutes per Goals 

##League Rank by Goals Conversion

```python
TOPGCleague2020=league2020.sort_values(by=['Goals_Conversion'],ascending=False)
sns.catplot(data=TOPGCleague2020,
            x='Goals_Conversion',
            y='League',
            kind='bar',
            ci=None,
            height=5, 
            aspect=2.2,)
```

Bundesliga is the league with highest Goals Conversion, Bundesliga has 54% Possibility to Score in Every Shot Target Occur.

##League Rank by Goals per Match Ratio

```python
TOPGPMleague2020=league2020.sort_values(by=['Goals_per_Match_Ratio'], ascending=False)
sns.catplot(data=TOPGPMleague2020,
            x='Goals_per_Match_Ratio',
            y='League',
            kind='bar',
            ci=None,
            height=5, 
            aspect=2.2,)
```

Bundesliga is the league with the highest goal per match ratio with 73% ratio. means in every match played in bundesliga, Possibility goals to occur is 73%.

##League Rank by Minutes per Goals

```python
TOPMPGleague2020=league2020.sort_values(by=['Minutes_per_Goals'])
sns.catplot(data=TOPMPGleague2020,
            x='Minutes_per_Goals',
            y='League',
            kind='bar',
            ci=None,
            height=5, 
            aspect=2.2,)
```

Serie A has a shortest minutes per goals ratio. Means Serie A is the league with longest time to score a goals. Meanwhile La Liga is league with longest  minutes per goals ratio. means La liga is the longest time to score a goal

##Rank League by Diff G - Xg

```python
TOPDGleague2020=league2020.sort_values(by=['Diff_GoalsXG'])
sns.catplot(data=TOPDGleague2020,
            x='Diff_GoalsXG',
            y='League',
            kind='bar',
            ci=None,
            height=5, 
            aspect=2.2,)
```

Serie A has the highest value of Diff Goals and XG. Means Serie A competition is much more productive than expected.  

#Team ,and Player most Productive in Bundesliga

```python
bundes2020=df2020[df2020['League']=='Bundesliga']
top20bundes2020=bundes2020.sort_values(by=['Goals'],ascending=False).reset_index().head(20)
top20bundes2020
```

```python
sns.catplot(data=top20bundes2020,
            x='Goals',
            y='Player Names',
            kind='bar',
            ci=None,
            height=5, 
            aspect=2.2,)
```

Robert Lewandowski, is the top scorer so far with 12 goals and 1,7 goals match . It makes him the most lethal scorer per Gameweek 9.

```python
top20bundes2020wrapped=top20bundes2020.groupby(by=['Club'])['Matches_Played',
       'Mins', 'Goals', 'xG', 'OnTarget',].sum().reset_index().sort_values(by=['Goals'],ascending=False)
top20bundes2020wrapped
```

```python
sns.catplot(data=top20bundes2020wrapped,
            x='Goals',
            y='Club',
            kind='bar',
            ci=None,
            height=5, 
            aspect=2.2,)
```

Bayern Munchen is the most productive teams with 20 goals scored in 23 matches they players played.

##Key Player of the most productive team in the league

```python
bayernkey2020=bundes2020[bundes2020['Club']=='(BAY)']
bayernkey2020=bayernkey2020.sort_values(by=['Goals'], ascending=False)
bayernkey2020
```

Robert Lewandowski is Key Player to makes Bayern Muenchen being the most productive teams in the league with his contribution to Score 12 Goals in 7 matches his played.
